```python
6. Write a while loop that starts with a variable counter = 5 and prints the counter, 
decreasing it by 1 each time, until it reaches 0 (print "Blastoff!" when it finishes).
```

```python
counter = 6

while (counter >= 1):
    counter -= 1
   
    print(counter)
print("Blastoff!")
```

```python
 7. Put it all together: Create a list called students. Inside this list, place three dictionaries. 
Each dictionary should represent a student and have two keys: "name" (a string) and "gpa" (a float).
-Write a for loop that iterates through the students list.
-Inside the loop, use an if statement to check if the student's GPA is greater than 3.0.
-If it is, print that student's name.
```

```python
students = [{"name" : "asfar",
            "gpa" : 3.83,},
            
            {"name" : "ahmad",
            "gpa" : 3.67},
            
            {"name" : "ali",
                "gpa" : 3.0}
            ]
for x in students:
    if x["gpa"] > 3.0:
        print(x)
        
```
